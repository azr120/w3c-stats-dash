# Totals per Division


```python
statistics_raw = df.groupby('ranking.leagueOrder')[['overallScore.HERO_SCORE', 'overallScore.UNIT_SCORE',
                              'resourceScore.GOLD_COLLECTED', 'resourceScore.LUMBER_COLLECTED', 'overallScore.RESOURCE_SCORE',
                           'overallScore.TOTAL_SCORE', 'updatedMmr.rating', ]].mean()
```

```python
division = {0.0: 'Grand Master',
            1.0: 'Master',
            2.0: 'Adept',
            3.0: 'Diamond',
            4.0: 'Platinum',
            5.0: 'Gold',
            6.0: 'Silver',
            7.0: 'Bronze',
            8.0: 'Grass'}
```

```python
m=(statistics_raw.dtypes=='float64')
statistics_raw.loc[:,m]=statistics_raw.loc[:,m].round().astype(int)
statistics_raw
```

```python
fig = go.Figure(data=[
    go.Bar(name='Hero Score', x=statistics_raw['League'], y=statistics_raw['Hero Score'], 
           text=statistics_raw['Hero Score'], textposition='auto'),
    go.Bar(name='Unit Score', x=statistics_raw['League'], y=statistics_raw['Unit Score'],
          text=statistics_raw['Unit Score'], textposition='auto'),
    go.Bar(name='Resource Score', x=statistics_raw['League'], y=statistics_raw['Resource Score'],
          text=statistics_raw['Resource Score'], textposition='auto'), 
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.update_layout(title="Hero, Unit & Resource Score per Division", xaxis_title="Divisions",
    yaxis_title="Score")
fig.show()
```

#  advanced per league, race 

```python
df['UNIT-RES-RATIO-PM'] = df['unitscorepm'].div(df['resourcescorepm'])
df['kill-participation'] = df['unitScore.UNITS_KILLED'] / (df['unitScore.UNITS_KILLED'] + df['e.unitScore.UNITS_KILLED'])
df['gold-lumber-ratioPM'] = df['gpm'].div(df['lpm'])
df['TSrelative'] = (df['overallScore.TOTAL_SCORE'] - df['e.overallScore.TOTAL_SCORE']) / df['e.overallScore.TOTAL_SCORE']
df['HeroSRelative'] = (df['overallScore.HERO_SCORE'] - df['e.overallScore.HERO_SCORE']) / df['e.overallScore.HERO_SCORE']
df['GoldRelative'] = (df['resourceScore.GOLD_COLLECTED'] - df['e.resourceScore.GOLD_COLLECTED']) / df['e.resourceScore.GOLD_COLLECTED']
df['UnitRelative'] = (df['overallScore.UNIT_SCORE'] - df['e.overallScore.UNIT_SCORE']) / df['e.overallScore.UNIT_SCORE']
```

```python
adv_statistics_pm_race = df.groupby(['ranking.leagueOrder'])[['UNIT-RES-RATIO-PM', 'kill-participation',
                                   'gold-lumber-ratioPM', 'heroScore.HEROES_KILLED', 'e.heroScore.HEROES_KILLED',
                                    'heroScore.MERCS_HIRED', 'updatedMmr.rating', 'duration']].mean()
```

```python
adv_statistics_pm_race = adv_statistics_pm_race.reset_index()
adv_statistics_pm_race['ranking.leagueOrder'] = adv_statistics_pm_race['ranking.leagueOrder'].map(division)

adv_statistics_pm_race.rename(columns={'ranking.leagueOrder' : 'League',
                                'UNIT-RES-RATIO-PM': 'Unit-res-ratio',
                                'kill-participation': 'Kill-Participation',
                                'gold-lumber-ratioPM': 'Gold-Lumber-Ratio',
                                'heroScore.HEROES_KILLED': 'HeroKills',
                                'e.heroScore.HEROES_KILLED': 'opponentHeroKills',
                                   'heroScore.MERCS_HIRED': 'Mercs',
                                   'updatedMmr.rating': 'Avg. MMR',
                             'duration': 'Avg. Duration'}, inplace=True)

adv_statistics_pm_race = adv_statistics_pm_race.round(2)
adv_statistics_pm_race['Avg. MMR'] = adv_statistics_pm_race['Avg. MMR'].astype('int')
adv_statistics_pm_race
```

# Player profiles

```python
players = pd.pivot_table(df, values=
                        [
                            '_id',
                            'heroscorepm', #pm
                            'unitscorepm',
                            'gpm',
                            'lpm',
                            'upkeeppm',
                            'resourcescorepm',
                            'totalscorepm',
                            'UNIT-RES-RATIO-PM',
                            'gold-lumber-ratioPM',
                            'duration',
                            'TSrelative', #rel
                            'HeroSRelative',
                            'GoldRelative',
                            'UnitRelative',
                            'kill-participation', #pg
                            'hero_combo',
                            'nheroes',
                            'heroScore.HEROES_KILLED', 
                            'e.heroScore.HEROES_KILLED',
                            'resourceScore.TECH_PERCENTAGE',
                            'heroScore.MERCS_HIRED',
                            'heroScore.ITEMS_OBTAINED',
                            'unitScore.LARGEST_ARMY',
                            'updatedMmr.rating',
                            'e.mmr.rating',
                            'ranking.leagueOrder',
                            'won',
                            'pwin',
                            'pings'
                        ], index=['battleTag', 'race', 'raceId'],
                        aggfunc= {
                            '_id': lambda x: len(x), 
                            'heroscorepm' : 'mean',
                            'unitscorepm' : 'mean',
                            'gpm': 'mean',
                            'lpm': 'mean',
                            'upkeeppm': 'mean',
                            'resourcescorepm': 'mean',
                            'totalscorepm': 'mean',
                            'UNIT-RES-RATIO-PM': 'mean',
                            'gold-lumber-ratioPM': 'mean',
                            'duration': [lambda x: x.sum()/60, 'mean'],
                            'TSrelative': 'mean', 
                            'HeroSRelative': 'mean',
                            'GoldRelative': 'mean',
                            'UnitRelative': 'mean',
                            'kill-participation': 'mean', 
                            'hero_combo': lambda x: x.mode().iat[0],
                            'nheroes': 'mean',
                            'heroScore.HEROES_KILLED': 'mean', 
                            'e.heroScore.HEROES_KILLED': 'mean',
                            'resourceScore.TECH_PERCENTAGE': 'mean',
                            'heroScore.MERCS_HIRED': 'mean',
                            'heroScore.ITEMS_OBTAINED': 'mean',
                            'unitScore.LARGEST_ARMY': 'mean',
                            'updatedMmr.rating': 'last',
                            'e.mmr.rating': 'mean',
                            'ranking.leagueOrder': 'last',
                            'won': [(lambda x: (x == True).sum()), (lambda x: (x == False).sum()),
                                          lambda x: (x == True).sum() / x.count()],
                            'pwin': 'mean',
                            'pings': 'mean'
                                })

players.columns = ['GoldVs', 'HeroVs', 'TotalVs', 'Unit-res-ratio', 'UnitVs', 'Games',
                  'Playtime', 'Duration', 'opponentHeroKills', 'opponentMMR', 'Gold-Lumber-Ratio',
                  'GPM', 'HeroKills', 'Items', 'Mercs', 'Hero_combo', 'HeroScorePM', 'Kill-Participation',
                  'LPM', 'No. heroes', 'Ping', 'Exp. Winrate', 'League', 'Tech Pct',
                  'ResourceScorePM', 'TotalScorePM', 'Largest Army', 'UnitScorePM',
                  'MMR', 'UpkeepPM', 'Won', 'Lost', 'Winrate']
```

```python
player_profiles = players[['Games', 'HeroScorePM', 'UnitScorePM', 'GPM', 'LPM', 'UpkeepPM', 'ResourceScorePM',
                          'TotalScorePM', 'Unit-res-ratio', 'Gold-Lumber-Ratio', 'Duration',
                          'TotalVs', 'HeroVs', 'GoldVs', 'UnitVs', 'Kill-Participation',
                          'Hero_combo', 'No. heroes', 'HeroKills', 'opponentHeroKills', 'Tech Pct',
                          'Mercs', 'Items', 'Largest Army', 'MMR', 'League', 'opponentMMR', 'Playtime',
                          'Won', 'Lost', 'Winrate', 'Exp. Winrate', 'Ping']].copy()
```

```python
player_profiles.to_csv('player_profiles.csv')
```

```python
player_profiles.reset_index(inplace=True)
```

```python
race2_mapping = {0.0: 'Random',
                1.0: 'HU',
                2.0: 'ORC',
                4.0: 'NE',
                8.0: 'UD'}

player_profiles['raceId'] = player_profiles['raceId'].map(race_mapping)
player_profiles['race'] = player_profiles['race'].map(race2_mapping)
```

```python
player_profiles['HeroScorePM'] = player_profiles['HeroScorePM'].round().astype(int)
player_profiles['UnitScorePM'] = player_profiles['UnitScorePM'].round().astype(int)
player_profiles['GPM'] = player_profiles['GPM'].round().astype(int)
player_profiles['LPM'] = player_profiles['LPM'].round().astype(int)
player_profiles['UpkeepPM'] = player_profiles['UpkeepPM'].round().astype(int)
player_profiles['ResourceScorePM'] = player_profiles['ResourceScorePM'].round().astype(int)
player_profiles['TotalScorePM'] = player_profiles['TotalScorePM'].round().astype(int)
player_profiles['Largest Army'] = player_profiles['Largest Army'].round().astype(int)

player_profiles['Ping'] = player_profiles['Ping'].round().astype(int) + 15


player_profiles['Gold-Lumber-Ratio'] = player_profiles['Gold-Lumber-Ratio'].round(2)
player_profiles['Unit-res-ratio'] = player_profiles['Unit-res-ratio'].round(2)
player_profiles['Duration'] = player_profiles['Duration'].round(2)
player_profiles['Kill-Participation'] = player_profiles['Kill-Participation'].round(2)
player_profiles['HeroKills'] = player_profiles['HeroKills'].round(2)
player_profiles['opponentHeroKills'] = player_profiles['opponentHeroKills'].round(2)
player_profiles['Tech Pct']= player_profiles['Tech Pct'].round(2)
player_profiles['Mercs'] = player_profiles['Mercs'].round(2)
player_profiles['Items'] = player_profiles['Items'].round(2)
player_profiles['MMR'] = player_profiles['MMR'].round(2)
player_profiles['opponentMMR'] = player_profiles['opponentMMR'].round(2)
player_profiles['Playtime'] = player_profiles['Playtime'].round(2)
player_profiles['Winrate'] = player_profiles['Winrate'].round(2)
player_profiles['Exp. Winrate'] = player_profiles['Exp. Winrate'].round(2)
player_profiles['TotalVs'] = player_profiles['TotalVs'].round(2)
player_profiles['HeroVs'] = player_profiles['HeroVs'].round(2)
player_profiles['GoldVs'] = player_profiles['GoldVs'].round(2)
player_profiles['UnitVs'] = player_profiles['UnitVs'].round(2)
player_profiles['No. heroes'] = player_profiles['No. heroes'].round(2)
```

```python
player_profiles = player_profiles.sort_values('MMR', ascending=False)
```

```python
player_profiles['League'] = player_profiles['League'].map(division)
```
